# EDA on Data

```python
import pandas as pd

# If needed, adjust path:
# df = pd.read_csv("/air_quality_health_dataset.csv")
df = pd.read_csv("air_quality_health_dataset.csv")

print("Shape:", df.shape)
print("\nColumns:\n", df.columns)

df.head(5)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

plt.hist(df["hospital_admissions"], bins=40)
plt.xlabel("Hospital admissions")
plt.ylabel("Count")
plt.title("Distribution of hospital admissions")
plt.show()
```

```python
plt.hist(df["temperature"], bins=40)
plt.xlabel("Temperature")
plt.ylabel("Count")
plt.title("Temperature distribution")
plt.show()

```

```python
plt.hist(df["aqi"], bins=40)
plt.xlabel("AQI")
plt.ylabel("Count")
plt.title("AQI distribution")
plt.show()
```

```python
plt.hist(df["pm2_5"], bins=40)
plt.xlabel("PM2.5")
plt.ylabel("Count")
plt.title("PM2.5 distribution")
plt.show()
```

```python
plt.scatter(df["temperature"], df["hospital_admissions"], alpha=0.05)
plt.xlabel("Temperature")
plt.ylabel("Hospital admissions")
plt.title("Temperature vs hospital admissions")
plt.show()
```

```python
plt.scatter(df["pm2_5"], df["hospital_admissions"], alpha=0.05)
plt.xlabel("PM2.5")
plt.ylabel("Hospital admissions")
plt.title("PM2.5 vs hospital admissions")
plt.show()
```

```python
import numpy as np

df["high_temp"] = df["temperature"] > df["temperature"].quantile(0.75)
df["high_pm"] = df["pm2_5"] > df["pm2_5"].quantile(0.75)

df.groupby(["high_temp", "high_pm"])["hospital_admissions"].mean()
```
